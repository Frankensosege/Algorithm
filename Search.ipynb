```python
# 순차 검색 : 자료가 정렬 되어있지 않은 상태에서 검색
def seqsearch(array, find_data):
    pos = []
    size = len(array)
    find = False
    
    for i in range(size):
        if array[i] == find_data:
            pos.append(i)
            find = True
    if not find:
        return None
    return pos # -1또는 검색된 자료의 인덱스
```

```python
def seqSearch(array, find_data):
    count = 0
    pos = -1
    size = len(array)
    
    for i in range(size):
        count += 1
        if array[i] == find_data:
            pos = i
            break
        elif array[i] > find_data:
            break
            
    return pos, count # -1또는 검색된 자료의 인덱스

def QuickSort(arr, start, end):
    if end <= start:
        return
    low = start
    high = end
#    pivot = arr[(low + high)//2] # (high - low) // 2 + low
    pivot = arr[(high - low) // 2 + low]
    while low <= high:
        while arr[low] < pivot:
            low += 1
        
        while arr[high] > pivot:
            high -= 1
        if low <= high:
            arr[low], arr[high] = arr[high], arr[low]
            low, high = low + 1, high - 1
    mid = low
    QuickSort(arr, start, mid-1)
    QuickSort(arr, mid, end)
    
array = [random.randrange(1, 1000) for _ in range(10)]
array_1 = random.choices(array, k=20)
array_sort = array_1.copy()
QuickSort(array_sort, 0, len(array_sort)-1)
print(array_1)
print(array_sort)
```

```python
def binary_search(arr, f_data):
    start = 0
    end = len(arr) - 1
    cnt = 0
    while start <= end:
        cnt += 1
        mid = (start + end) // 2
        if arr[mid] == f_data:
            return mid, cnt
        elif arr[mid] > f_data:
            end = mid -1            
        else:
            start = mid + 1
    return -1, cnt

array = [random.randrange(1, 100000) for _ in range(100000)]
# QuickSort(array, 0, len(array)-1)
array.sort()

find_data = random.randrange(1, 100000)

pos, cnt = binary_search(array, find_data)

print(f"배열 일부 --> {array[0:5]}, ....., {array[len(array) - 6:len(array) - 1]}")
if pos == -1:
    print(f"{find_data} 검색된 자료 없음....{cnt}번 검색")
else:
    print(f"{find_data} 는 {pos}번째에.. {cnt}번 검색")
```

```python
# 1. 책장배열을 정의한다.
bookAry = [ ['어린왕자', '쌩떽쥐베리'],['이방인', '까뮈'], ['부활', '톨스토이'],
            ['신곡', '단테'], ['돈키호테', '세르반테스'], ['동물농장', '조지오웰'],
            ['데미안','헤르만헤세'], ['파우스트', '괴테'], ['대지', '펄벅'] ]
# class 자료로 변경
# 2. 도서명만 추출, 도서명 옆에 책의 순번(책장의 위치)를 출력
#    -> make_idx_auth(), mak_idx_book()
# 3. 도서명을 기준으로 색인표 정렬 -> sorted 함수로 0번째 열인 도서명 열을 기준으로 -> sorted() 정렬된 자료 리스트 반환
```

```python
# 작가명으로 색인표 생성 출력 -> 각각의 데이터는 클래스로
class Book:
    def __init__(self, title, auth):
        self.auth, self.title = auth, title
    
    def __repr__(self):
        return repr((self.title, self.auth))
    
def make_idx(arr):
    books = []
    for book in arr:
        books.append(Book(book[0], book[1]))
        
    return books

def sort_books(books, key='title'):
    if key == 'title':
        return sorted(books, key=lambda book:book.title)
    else:
        return sorted(books, key=lambda book:book.auth)

def search_book(arr, start, end, f_data):
    start = 0
    end = len(arr) - 1
    while start <= end:
        mid = (start + end) // 2
        if arr[mid].title == f_data:
            return mid
        elif arr[mid].title  > f_data:
            end = mid -1            
        else:
            start = mid + 1
    return None

def print_books(books, sw=None):
    print()   
    if sw:
        print("----  정렬된 색인표를 생성한 책장 ---- ")
    else:
        print("---- 색인표를 생성한 책장 ---- ")
    for idx, book in enumerate(books):     # 인덱스 생성된 책장 정보 출력
        print(f'{idx} : {repr(book)}')
        
books = make_idx(bookAry)
# print("------ 색인표 생성 책장 ------")
# for idx, book in enumerate(books):
#     print(f"{idx} : {repr(book)}")
print_books(books)
# 도서명으로 색인표 생성 출력
srt_books = sort_books(books, key='title')
print("------ 색인표 생성 책장 ------")
for idx, book in enumerate(srt_books):
    print(f"{idx} : {repr(book)}")
# 도서명을 기준으로 색인표 정렬
srt_books_1 = sorted(books, key=lambda book:book.title)
print_books(srt_books_1, sw=True)   #  정렬된 자료 출력
# 검색하고자 하는 도서명을 입력 -> 결과 출력
f_title = input("검색 하고자 하는 도서명 입력 -> ")
pos = search_book(srt_books_1, 0, len(srt_books_1) - 1,  f_title)   # 검색된 위치 반환
if pos == -1:
    print("검색할 자료 없음 ")
else:
    print("{} 번째 위치한 책의 정보 {}".format(pos, repr(srt_books_1[pos])))
```
